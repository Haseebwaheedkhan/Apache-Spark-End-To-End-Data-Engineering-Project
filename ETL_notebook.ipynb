# Factory Patter


```python
class Person:
    person_count = 0

    def __init__(self, p_name, p_age):
        self.name = p_name
        self.age = p_age
        Person.person_count +=1
    
    def show_details(self):
        print(f"My name is : {self.name} and my age is {self.age}")


p = Person("Haseeb", 56)
p.show_details()

p.person_count
```
